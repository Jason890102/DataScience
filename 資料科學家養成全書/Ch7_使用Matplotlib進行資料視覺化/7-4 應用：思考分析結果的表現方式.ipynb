```python
# 指定有著資料的url
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip'

# 從url取得資料
r = requests.get(url,stream=True)

# 讀取zipfile並在當前目錄展開檔案
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

##### 2.移除$記號、把data變換為date型別

```python
# 型別變更 日期時間型別
dow_jones_index_data.date = pd.to_datetime(dow_jones_index_data.date)

# 消除$符號
delete_dolchar = lambda x: str(x).replace('$', '')

# 對象為open, high, low, close, next_weeks_open, next_weeks_close
# 將字串型別轉換為數值型別的處理，因前面已將數值轉為字串
dow_jones_index_data.open = pd.to_numeric(dow_jones_index_data.open.map(delete_dolchar))
dow_jones_index_data.high = pd.to_numeric(dow_jones_index_data.high.map(delete_dolchar))
dow_jones_index_data.low = pd.to_numeric(dow_jones_index_data.low.map(delete_dolchar))
dow_jones_index_data.close = pd.to_numeric(dow_jones_index_data.close.map(delete_dolchar))
dow_jones_index_data.next_weeks_open = pd.to_numeric(dow_jones_index_data.next_weeks_open.map(delete_dolchar))
dow_jones_index_data.next_weeks_close = pd.to_numeric(dow_jones_index_data.next_weeks_close.map(delete_dolchar))

# 再次確認
dow_jones_index_data.head()
```

```python
dow_jones_index_data
```

##### 3

```python
# 設定index
dow_jones_index_data_stock_index = dow_jones_index_data.set_index(['date','stock']) # 將stock設為index
# 為何要用date？

# DataFrame的重新組成
dow_jones_index_data_stock_index_unstack = dow_jones_index_data_stock_index.unstack()
# 把列轉成行，原本stock的資料是列，若直接做describe會直接被覆蓋過去，變成750筆資料一起算，所以需要轉成行來計算

# 對close做摘要統計量
dow_close_data = dow_jones_index_data_stock_index_unstack['close']
dow_close_data.describe()
```

##### 4

```python
# stock之間的相關矩陣
corr_data = dow_jones_index_data_stock_index_unstack['close'].corr()
corr_data
```

```python
sns.heatmap(corr_data)
```

##### 5.

```python
# 還未看懂QQ
# initial value
max_corr = 0
stock_1 = ''
stock_2 = ''

for i in range(0,len(corr_data)):
    print(
        corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[1], # sort_values為降序排列、idxmax為找出最大值的索引
        corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[0],
        corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]
    )
    if max_corr < corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]:
        max_corr = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]
        stock_1 = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[1]
        stock_2 = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[0]

# 輸出max_corr的Pair
print('[Max Corr]:',max_corr)
print('[stock_1]',stock_1) # 相關係數最高的組合
print('[stock_2]',stock_2) # 相關係數最高的一對
```

##### 7.

```python
# 想取得前週比（錯開1期），可使用shift
# 相較於使用loop等，速度絕對快很多
log_ratio_stock_close = np.log(dow_close_data/dow_close_data.shift(1))

max_vol_stock = log_ratio_stock_close.std().idxmax() # 取最大標準差
min_vol_stock = log_ratio_stock_close.std().idxmin() # 取最小標準差

# 最大值與最小值的標準差之stock
print('max volatility:',max_vol_stock)
print('min volatility:',min_vol_stock)

# 圖形化
log_ratio_stock_close[max_vol_stock].plot()
log_ratio_stock_close[min_vol_stock].plot()
plt.ylabel('log ratio')
plt.legend()
plt.grid(True)
```

##### 綜合問題7-2 市場行銷分析

```python
online_retail_data_table = pd.ExcelFile('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx').parse('Online Retail') # parse為指定工作表
online_retail_data_table.head()
```

```python
# 確認資料
online_retail_data_table.info()
```

```python
# 抽出InvoiceNo第一個字元放置在最右側cancel_fig之行，使用map及lambda之函式
online_retail_data_table['cancel_fig'] = online_retail_data_table.InvoiceNo.map(lambda x: str(x)[0])
online_retail_data_table.groupby('cancel_fig').size()
```

```python
# 由於「C」開頭的是取消的資料，撰寫去除的處理
# 「A」也視為異常值處理刪除
# 從上述結果得知，這次只將開頭為「5」的是為分析對象
# 此外，只有將CustomerID的資料視為對象
online_retail_data_table = online_retail_data_table[(online_retail_data_table.cancel_fig == '5') & (online_retail_data_table.CustomerID.notnull())]
```

##### 2.

```python
# unique ID
print('購買者數量(Unique):',len(online_retail_data_table.CustomerID.unique()))

# unique StockCode
print('商品紀錄數量:',len(online_retail_data_table.StockCode.unique()))

# unique description
print('商品名稱的種類數量:',len(online_retail_data_table.Description.unique()))

# unique bascket
print('購物籃數量:',len(online_retail_data_table.InvoiceNo.unique()))
```

##### 3.

```python
# 計算總額，放置在新增欄位
online_retail_data_table['Totalprice'] = online_retail_data_table.Quantity * online_retail_data_table.UnitPrice

# 對各個國家計算銷售總和金額
country_data_total_p = online_retail_data_table.groupby('Country')['Totalprice'].sum()

# 對於銷售總和金額從大排到小，取出前五
top_five_country = country_data_total_p.sort_values(ascending = False)[0:5]

# TOP5的國家
print(top_five_country)

# TOP5的國家List
print('TOP5國家的List:',top_five_country.index)
```

##### 4.

```python
# 製作只有TOP5的資料
top_five_country_data = online_retail_data_table[online_retail_data_table['Country'].isin(top_five_country.index)]

# date與別國的銷售
top_five_country_data_country_totalP = top_five_country_data.groupby(['InvoiceDate','Country'],as_index = False)['Totalprice'].sum()

# TOP5的銷售之每月時間序列推移
# index的設定(日期時間與國家)
top_five_country_data_country_totalP_index = top_five_country_data_country_totalP.set_index(['InvoiceDate','Country'])

# 重新組成
top_five_country_data_country_totalP_index_uns = top_five_country_data_country_totalP_index.unstack()

# 使用resamplt能將時間序列資料變更為每月或每季度。這次計算每月(M)的總和。之後進行圖形化
top_five_country_data_country_totalP_index_uns.resample('M').sum().plot(subplots = True, figsize = (12,10))

# 使得圖形不至於重疊
plt.tight_layout()
```

##### 5.

```python
for x in top_five_country.index:
    country = online_retail_data_table[online_retail_data_table['Country'] == x]
    country_stock_data = country.groupby('Description')['Totalprice'].sum()
    top_five_country_stock_data = pd.DataFrame(country_stock_data.sort_values(ascending=False)[0:5])
    plt.figure()
    plt.pie(
        top_five_country_stock_data,
        labels = top_five_country_stock_data.index,
        counterclock = False,
        startangle = 90,
        autopct = '%.1f%%',
        pctdistance = .7
    )
    plt.ylabel(x)
    plt.axis('equal')
```
