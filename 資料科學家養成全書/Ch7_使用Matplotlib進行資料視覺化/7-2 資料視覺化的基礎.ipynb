```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series,DataFrame

# 視覺化函式庫
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

# 顯示到小數點後第3位
%precision 3
```

```python
# 想依照某個類別來對數值進行比較時，會使用長條圖

# 顯示的資料
x = [1,2,3]
y = [10,1,4]

# 指定圖形的大小
plt.figure(figsize = (10,6))

# align='center'為讓圖形置中
plt.bar(x,y,align='center',width = .5)

# 長條圖的各個標籤
plt.xticks(x,['A Class','B Class','C Class'])

# 設定x與y的標籤
plt.xlabel('Class')
plt.ylabel('Score')

# 顯示格線
plt.grid(True)
```

```python
# 顯示的資料
x = [1,2,3]
y = [10,1,4]

# 指定圖形的大小
plt.figure(figsize = (10,6))

plt.barh(x,y,align='center')
plt.yticks(x,['A Class','B Class','C Class'])
plt.xlabel('Score')
plt.ylabel('Class')
plt.grid(True)
```

```python
# 描繪多個圖形

# 資料準備
y1 = np.array([30,10,40])
y2 = np.array([10,50,90])

# X軸資料
x = np.arange(len(y1))

# 圖形寬度
w = .4

# 指定圖形大小
plt.figure(figsize = (10,6))

# 描繪圖形，對於y2，將其右移一個圖形寬度來描繪
plt.bar(x, y1, color = 'blue', width = w, label = 'Math first', align = 'center')
plt.bar(x + w, y2, color = 'green', width = w, label = 'Math final', align = 'center')

# 讓圖例配置於最適當的位置
plt.legend(loc = 'best')

plt.xticks(x + w/2, ['Class A','Class B','Class C'])
plt.grid(True)
```

```python
# 堆疊長條圖

height1 = np.array([100,200,300,400,500])
height2 = np.array([1000,800,600,400,200])
# X軸
x = np.array([1,2,3,4,5])

# 指定圖形的大小
plt.figure(figsize = (10,6))

# 圖形描繪
p1 = plt.bar(x, height1, color = 'blue')
p2 = plt.bar(x, height2, bottom = height1, color = 'lightblue')

# 顯示圖例
plt.legend((p1[0],p2[0]),('Class 1','Class 2'))
```

```python
# 圓餅圖

labels = ['Frogs','Hogs','Dogs','Logs']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,.1,0,0)

# 指定圖形的大小
plt.figure(figsize = (15,6))

# 顯示圖形
plt.pie(sizes, explode = explode, labels = labels, colors = colors, # explode指脫離全體的圓形圖
        autopct = '%1.1f%%', shadow = True, startangle = 90) # autopct指定用來顯示各自所佔比例的格式

# 維持比例描繪圖形
plt.axis('equal')
```

```python
# 泡泡圖

N = 25

# 隨機生成x, y資料
x = np.random.rand(N)
y = np.random.rand(N)

# color編號
colors = np.random.rand(N)

# 讓泡泡的大小顯得散亂
area = 10 * np.pi * (15 * np.random.rand(N)) ** 2

# 指定圖形大小
plt.figure(figsize = (15,6))

# 描繪圖形
plt.scatter(x, y, s = area, c = colors, alpha = .5)
plt.grid(True)
```
